```python
import matplotlib.pyplot as plt

# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Customize the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=5, edgecolor='black', color='skyblue')  # Create the histogram
plt.title('Histogram of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.xticks(range(1, 6))  # Set specific x-axis ticks
plt.grid(axis='y', alpha=0.75)  # Add a horizontal grid for better readability

# Display the histogram
plt.show()
```
