```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the subplots
# Line chart data
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

# Bar chart data
categories_bar = ['X', 'Y', 'Z']
values_bar = [15, 30, 22]

# Scatter plot data
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Pie chart data
labels_pie = ['Part 1', 'Part 2', 'Part 3']
sizes_pie = [45, 30, 25]
colors_pie = ['lightcoral', 'lightskyblue', 'lightgreen']

# Create the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Combined Subplots')  # Add a title for the entire figure

# First subplot: Line chart
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('sin(X)')
axs[0, 0].grid(True)

# Second subplot: Bar chart
axs[0, 1].bar(categories_bar, values_bar)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')

# Third subplot: Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', startangle=90)
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle

# Display the plots
plt.show()
```
